```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
plt.style.use("ggplot")
```

```python
X = np.sort(5*np.random.rand(200,1),axis = 0)
X.shape 
```

```python
Y = np.sin(X).ravel()
Y[::5] += 3*(0.5-np.random.rand(40)) # Agregar Ruido
```

```python
plt.figure(figsize = (16,6))
plt.scatter(X,Y, color = "darkorange", label = "data")
plt.legend()
plt.show()
```

```python
C = 1e3
sv_linear = SVR(kernel = "linear", C = C)
sv_rbf = SVR(kernel = "rbf", gamma = 0.01, C = C)
sv_poly = SVR(kernel = "poly", C = C, degree=3)
```

```python
y_lin = sv_linear.fit(X,Y).predict(X)
y_rbf = sv_rbf.fit(X,Y).predict(X)
y_poly = sv_poly.fit(X,Y).predict(X)
```

```python
lw = 2
plt.figure(figsize = (20,9))
plt.scatter(X,Y,color = "darkorange",lw = 2, label = "Data")
plt.plot(X,y_lin,color = "navy",lw = 2, label = "Kernel Linear")
plt.plot(X,y_rbf,color = "cyan",lw = 2, label = "Kernel Radial")
plt.plot(X,y_poly,color = "black",lw = 2, label = "Kernel Poly")
plt.xlabel("X")
plt.xlabel("Y")
plt.title("Support Vector Regression")
plt.legend()
plt.show()
```

```python
sv_rbf.score(X,Y),sv_linear.score(X,Y), sv_poly.score(X,Y)
```
