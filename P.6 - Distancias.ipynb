### Minkowski distance

The Minkowski distance is a metric in a normed vector space which can be considered as a generalization of both the Euclidean distance and the Manhattan distance.

Definition

$$X = (x_1,x_2,\cdots,x_n)\ \ \ and\ \  Y\ = (y_1,y_2,\cdots,y_n)$$
$$D(X,Y) = \left( \sum_{i=1}^n |x_i-y_i|^p \right){1/P}$$

```python
dd1 = distance_matrix(data,data,p=1)
dd1
```

```python
dd2 = distance_matrix(data,data,p=2) # Distancia Euclidea
dd2
```

## Clustering Jerárquico

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
movies
```

```python
data[movies]
```

```python
z = linkage(data[movies], method="complete",metric="euclidean")
# The distance function can be ‘braycurtis’, ‘canberra’, ‘chebyshev’, ‘cityblock’, 
#‘correlation’, ‘cosine’, ‘dice’, ‘euclidean’, ‘hamming’, ‘jaccard’, ‘jensenshannon’, ‘kulsinski’,
#‘mahalanobis’, ‘matching’, ‘minkowski’, ‘rogerstanimoto’, ‘russellrao’, ‘seuclidean’, 
# ‘sokalmichener’, ‘sokalsneath’, ‘sqeuclidean’, ‘yule’.
z
```

```python
plt.figure(figsize=(10,8),facecolor="lightblue" )
plt.title("Dendograma Jerárquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(z,leaf_rotation = 90, leaf_font_size=10)
plt.show()
```
